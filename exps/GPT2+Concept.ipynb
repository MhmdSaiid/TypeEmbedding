```python
import torch
import numpy as np
```

```python
from transformers import GPT2Tokenizer, GPT2LMHeadModel
tokenizer = GPT2Tokenizer.from_pretrained('gpt2')
device = torch.device('cuda') if torch.cuda.is_available() else torch.device('cpu')
model = GPT2LMHeadModel.from_pretrained('gpt2', pad_token_id=tokenizer.eos_token_id)
vocab = list(tokenizer.get_vocab().keys())
```

```python
def forward_pre_hook(vector2add,k=1):
    def pre_hook(module, input):
        #print(input[0].shape)
        for i in range(len(input[0][0])):
            input[0][0][i] = input[0][0][i] + k*vector2add
        return input
    return pre_hook
```

```python
phook=None
```

```python
if phook: phook.remove()
#phook = model.transformer.drop.register_forward_pre_hook(forward_pre_hook(first,0))

from transformers import pipeline, set_seed
generator = pipeline('text-generation', model=model,tokenizer=tokenizer)
set_seed(0)
generator("I like", max_length=1000, num_return_sequences=1)[0]['generated_text']
```

```python
locs =['France','China','Italy','ĠItaly','ĠFrance','ĠIreland']
loc_indices=[vocab.index(n) for n in locs]
with torch.no_grad():
    vec = torch.stack([model.transformer.wte(torch.tensor(i).to(device)) for i in loc_indices])
u,s,vh = np.linalg.svd(vec.cpu(),full_matrices=True)
first = torch.tensor(-vh[0])
```

```python
phook.remove()
phook = model.transformer.drop.register_forward_pre_hook(forward_pre_hook(first,2))
from transformers import pipeline, set_seed
generator = pipeline('text-generation', model=model,tokenizer=tokenizer)
set_seed(0)
generator("I like", max_length=1000, num_return_sequences=1)[0]['generated_text']
```
